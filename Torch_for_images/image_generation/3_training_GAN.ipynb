#### **Building a GAN**

First, let us define Loss functions for the model

#### **Generator Objective**



#### **Defining loss function for Generator**



```python
import torch
import torch.nn as nn

# Generator Loss Function
def gen_loss(gen, disc, num_images, z_dim):
    noise = torch.randn(num_images, z_dim)
    fake = gen(noise)   # generator output image
    disc_pred = disc(fake)  # discriminator prediction (fake / real)
    criterion = nn.BCEWithLogitsLoss()  
    gen_loss = criterion(disc_pred, torch.ones_like(disc_pred))
    return gen_loss
```

#### **Discriminator Objective**



#### **Defining Loss Function for Discriminator**



```python
# Discriminator Loss

def disc_loss(gen, disc, real, num_images, z_dim):
    criterion = nn.BCEWithLogitsLoss(num_images, z_dim)
    
    # Loss for fake (generator produced) images
    fake = gen(noise)
    disc_pred_fake = disc(fake)
    fake_loss = criterion(disc_pred_fake, torch.zeros_liks(disc_pred_fake))
    
    # Loss for real images
    disc_pred_real = disc(real)
    real_loss = criterion(disc_pred_realm torch.ones_like(disc_pred_real))
    
    # Average loss
    disc_loss = (real_loss + fake_loss) / 2
    return disc_loss
```
