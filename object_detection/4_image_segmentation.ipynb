#### **Data Annotations for Segmentations**

```python
from PIL import Image
from torchvision import transforms

image = Image.open('images/british_cat.jpg')
mask = Image.open('annots/british_cat.png')

transform = transforms.Compose([transforms.ToTensor()])

image_tensor = transform(image)
mask_tensor = transform(mask)

print(f'Image Shape: {image_tensor}\nMask Shape: {mask_tensor}')
```

output - 



image used -



From dataset documentation, we can see that \
Pixel annotations can only take 3 values,
- 1: Foreground (object)
- 2: Background
- 3: Not Classified

#### **Creating a binary mask**

```python
import torch
import matplotlib.pyplot as plt

binary_mask =  torch.where(
    mask_tensor == 1/255,
    torch.tensor(1.0),
    torch.tensor(0.0),
)

to_pil_image = transforms.ToPILImage()
mask = to_pil_image(binary_mask)
plt.imshow(mask)
```

#### **Segmenting the object**

```python
object_tensor = image_tensor * binary_mask

to_pil_image = transforms.ToPILImage()
object_image = to_pil_image(object_tensor)

plt.imshow(object_image)
```
