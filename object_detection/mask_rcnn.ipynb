#### **Pre-Trained masked R-CNN in PyTorch**



```python
from PIL import Image
from torchvision.models.detection import maskrcnn_resenet50_fpn  # Trained on COCO dataset
from torchvision import transforms

model = maskrcnn_resenet50_fpn(pretrained=True)
model.eval()

image = Image.open('cat_and_laptop.jpg')
transform = transforms.Compose([transforms.ToTensor()])
image_tensor = transform(image).unsqueeze(0)

with torch.no_grad():
    prediction = model(image_tensor)
```
