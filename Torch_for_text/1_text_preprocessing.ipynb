#### **Tokenization**

- Tokens can be a full word, part of a word or a punchuation. They are extracted from the text.
- Tokenization is using `torchtext` library

```python
from torchtext.data.utils import get_tokenizer

tokenizer = get_tokenizer('basic_english')
tokens = tokenizer('I am reading a book now. I love to read books!')
print(tokens)
```
