#### **Adversarial Attacks on Text Classification Models**

Ans) An adversarial attack includes tweaking the input data to encorporate malicious changes into it that can drastically affext AI's decision-making.

#### **FGSM - Fast Gradient Sign Method**

It exploits the model's learning information and makes the tiniest possible change to deceive the model.



#### **PGD - Projected Gradeint Descent**

It is more advanced than FGSM because it is iterative. It tries to find the most effective disturbance by iteratively modifying the input data.



#### **C&W Attack - The Carlini & Wagner Attack**

It focuses on optimizing the loss function. It ensures that the modifications are not just about deceiving but are being undetectable.





#### **Defence Strategies**

- Using Model Ensembling, i.e. Using multiple models



- Using Robust Data Augmentation
- Using Adversarial Training that anticipates deception

#### **Defence Tools**

- Using PyTorch's `Robustness Toolbox` to strengthen text models
- Using Gradient Masking by adding a variety of data to hide exploitable patterns
- Implementing regularization techniques to ensure model balance
